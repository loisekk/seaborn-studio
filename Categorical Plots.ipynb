# Seaborn –  Categorical Plots


## What Are Categorical Plots?

Categorical plots visualize data where at least one variable is **categorical**  
(e.g., day of week, gender, class).

Seaborn gives several easy options to:
- Compare **counts** or **averages** across categories
- Understand spread & distribution


## barplot() vs countplot()

```python
import seaborn as sn
import matplotlib.pyplot as plt


tips = sn.load_dataset('tips')
tips.head()

# bar plot 
sn.barplot(x = 'day' , y = 'total_bill' , data = tips , color = 'purple' )
```

## boxplot() & violinplot()

```python
sn.boxenplot(x = 'total_bill' , y = 'day' , hue = 'time' , data = tips , palette= 'rainbow') # box plot 
sn.boxplot(x = 'total_bill' , y = 'day' , hue = 'time' , data = tips ,  palette= 'rainbow') # box plot 

sn.violinplot(x = 'total_bill' , y = 'day' , hue = 'time' , data = tips ,  palette= 'rainbow')
```

## stripplot() & swarmplot()

```python
sn.stripplot(x = 'day' , y = 'total_bill' , data = tips , color = 'red' )
```

```python
# Swarmplot 

sn.swarmplot(x = 'day' , y = 'total_bill' , data = tips , color = 'red')
```

```python
sn.stripplot(x = 'day' , y = 'total_bill' , hue = 'time' , data = tips , color = 'red'  )
plt.grid(True)
sn.swarmplot(x = 'day' , y = 'total_bill' , hue  = 'time' , data = tips , color = 'red')
```

| Plot        | Best for |
|-------------|-----------|
| **barplot** | Compare average values across categories |
| **countplot** | Show how many items fall in each category |
| **boxplot** | See median, quartiles, and outliers |
| **violinplot** | Same as boxplot + shows distribution |
| **stripplot** | View every observation (good for small data) |
| **swarmplot** | Like stripplot but avoids overlap |
