# Seaborn –  Regression & Mixed Plots

## Regression & Mixed Plots

Seaborn provides plots that combine **relationships** and **statistical models** or mix multiple plot types.

- `regplot()` / `lmplot()` → scatter + regression line  
- `jointplot()` → scatter with hist/KDE  
- `pairplot()` → multiple relationships in one go


# regplot() – Scatter + Regression Line

```python
import seaborn as sn
import matplotlib.pyplot as plt


df = sn.load_dataset('tips')
df
df.head()
sn.regplot(x = 'total_bill' , y = 'tip' , data = df , robust=True , scatter = True , truncate= bool , marker = 'o' )
```

# lmplot() – Regression with Facets

```python
sn.lmplot(x = 'total_bill' , y = 'tip' , data = df , palette= 'rainboow'    )
```

# jointplot() – Scatter + Histogram / KDE

```python
sn.jointplot(x  = 'total_bill' , y = 'tip' , data = df , kind = 'reg' , palette= 'rainbow' )
```

# pairplot() – Relationships Between Many Variables

```python
sn.pairplot(df , hue = 'time' , palette= 'rainbow')
```

```python
iris = sn.load_dataset('iris')
iris.head()
```

```python
sn.pairplot(data = iris , hue = 'species' , palette='rainbow')
```
